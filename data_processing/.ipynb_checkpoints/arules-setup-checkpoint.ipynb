```python
# now produce an array representing the length category for each review
#len_categories = np.zeros(176504, dtype=str)
len_categories = []
infile = "drug_data_trimmed.csv"
# regenerate non-sorted review_lengths
review_lengths = np.zeros(176504, dtype=float)
with open(infile, "r") as f:
    fcsv = csv.DictReader(f)
    for i, line in enumerate(fcsv):
        rev_len = len([char for char in line["review"] if char.isalpha()])
        review_lengths[i] = rev_len
        if rev_len < 166:
            #len_categories[i] = "v.short"
            len_categories.append("v.short")
        elif 166 <= rev_len < 282:
            #len_categories[i] = "short"
            len_categories.append("short")
        elif 282 <= rev_len < 410:
            #len_categories[i] = "medium"
            len_categories.append("medium")
        elif 410 <= rev_len < 548:
            #len_categories[i] = "long"
            len_categories.append("long")
        else:
            #len_categories[i] = "v.long"
            len_categories.append("v.long")
```

```python
outfile = "drug_data_trimmed_text_len.csv"
infile = "drug_data_trimmed.csv"
with open(infile, "r") as f, open(outfile, "w") as o:
    fcsv = csv.DictReader(f)
    ocsv = csv.writer(o)
    ocsv.writerow(fcsv.fieldnames + ["revLength"] + ["revLenDesc"])
    for i, line in enumerate(fcsv):
        newline = list(line.values()) + [review_lengths[i]] + [len_categories[i]]
        ocsv.writerow(newline)
```
