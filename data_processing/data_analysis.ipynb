```python
import matplotlib.pyplot as plt
```

```python
# this data represents the average silhouette width using kmedoids clustering
# for the 4 fingerprints with cluster sizes 25-250 (incremented by 25)
mystr = "0.08746576 0.1125489 0.1455389 0.1719685 0.1967943 0.2229049 0.2473036 0.2735536 0.2935038 0.3153490 0.32861879 0.4059009 0.4569925 0.5101281 0.5549845 0.5899273 0.6228682 0.6529827 0.6740746 0.6982324 0.11164533 0.1284618 0.1534368 0.1761203 0.1926051 0.2084645 0.2223111 0.2457372 0.2606693 0.2789686 0.09786234 0.1161481 0.1311379 0.1535196 0.1712941 0.1911093 0.2103453 0.2276654 0.2397479 0.2577854"
moo = mystr.split(" ")
for item in moo:
    print(item, end=", ")
```

```python
fp2 = [float(x) for x in moo[0:10]]
fp3 = [float(x) for x in moo[10:20]]
fp4 = [float(x) for x in moo[20:30]]
maccs = [float(x) for x in moo[30:]]
numclusters = [x for x in range(25, 251, 25)]
#plt.figure(figsize=(8,6))
plt.plot(numclusters, fp2, marker='o', color='blue', linewidth=2, label="FP2")
plt.plot(numclusters, fp3, marker='*', color='red', linewidth=2, label="FP3")
plt.plot(numclusters, fp4, marker='v', color='orange', linewidth=2, label="FP4")
plt.plot(numclusters, maccs, marker="", color="green", linewidth=2, label="MACCS")
plt.legend()
plt.ylabel("Average Silhouette Width")
plt.xlabel("Number of Kmedoids Clusters")
#plt.show()
plt.savefig("silwidths_kmed.png", dpi=400)
```

```python
assert len(fp2) == len(fp3) == len(fp4) == len(maccs)
```
